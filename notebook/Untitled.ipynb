```python
img1 = cv2.imread('C:/Users/chau/Desktop/ShoeSizeProject/Images/train/24032.jpeg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img1 = cv2.resize(img1, (512, 768))
```

```python
plt.imshow(img1)
```

```python
edges = cv2.Canny(img1, 50, 150)
plt.imshow(edges)
```

```python
import numpy as np
kernel = np.ones((1,5), np.uint8)
img_erosion = cv2.erode(edges, kernel, iterations = 1)
img_dilate = cv2.dilate(img_erosion, kernel, iterations = 1)
```

```python
fig, ax = plt.subplots(1,2)
ax[0].imshow(img_erosion)
ax[1].imshow(img_dilate)
plt.show()
```

```python
kernel_ = np.ones((3,1), np.int8)
img_erosion_ = cv2.erode(edges, kernel_, iterations = 1)
img_dilate_ = cv2.dilate(img_erosion_, kernel_, iterations = 1)
```

```python
fig, ax = plt.subplots(1,2)
ax[0].imshow(img_erosion_)
ax[1].imshow(img_dilate_)
plt.show()
```
